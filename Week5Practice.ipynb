A codon is a set of three RNA bases that code for a single amino acid or indicate the end of a reading frame. The stop codons are UAA, UAG, and UGA.

Search the given dictionary of DNA sequences for stop codons and print the starting index of each stop codon. Since codons are always three bases long, you will need to iterate over each sequence such that you check if there is a stop codon every three bases, and only read three successive bases. Assume that the first letter is the start of the first codon.

In the example sequence "AGCUCGUAA", there are three codons, "AGC", "UCG", and "UAA". While the sequence "GCU" is found in this larger sequence, GCU is not a codon because it starts on the second base in the sequence. A sequence's codons will always be three bases long, with the first codon starting at the beginning of the sequence.

```python
#Activity 4 Completion Space
import re
sequenceDict = {"seq1":"AAAUAAGCGGCG","seq2":"ACGGCGUAGGCG","seq3":"GGCCUAUCCUGA"}

#P.S. The pipe operator, | , can be used inside of regular expressions to mean "or".


```

Click "Show code" below to see a sample solution to the activity.

```python
# @title Activity 4 Solution { display-mode: "form" }
import re
sequenceDict = {"seq1":"AAAUAAGCGGCG","seq2":"ACGGCGUAGGCG","seq3":"GGCCUAUCCUGA"}

codon = "" #Create a variable to hold the codon you're looking at at a given time

for sequence in sequenceDict.values(): #Iterate through the dictionary values
  counterStart = 0 #Create counters for the beginning and end of each codon
  counterStop = 3 #This is set to 3 because we will use list slicing, where the character at the index of the second slice value is not included in the slice
  while counterStop <= len(sequence): #Iterate through each sequence
    codon = sequence[counterStart:counterStop]  #Get the codon with list slicing
    if (re.search(r"UA[AG]{1}|UGA",codon)): #Check using a regular expression for each possible codon. The | character is an OR operator.
      print("Stop codon found at index", counterStart)
    else:
      pass
    counterStart = counterStart + 3 #Increase both counters by three so you start at the next codon
    counterStop = counterStop + 3
```
