## Activity 2

Using its AphiaID number, query the World Register of Marine Species API and print all non-English common names for Crassostrea virginica.

The root url for all queries is https://www.marinespecies.org/rest .

The API application to obtain an AphiaID is /AphiaIDByName/{ScientificName}

The API applicaiton to obtain common names using and AphiaID is /AphiaVernacularsByAphiaID/{aphiaID}

```python
#Activity 2 Completion Space

```

Click "Show code" below to see a sample solution to the activity.

```python
# @title Activity 2 Solution { display-mode: "form" }

import requests
from pprint import pprint

sciname = "Crassostrea virginica"
aphiaID = requests.get(f"https://www.marinespecies.org/rest/AphiaIDByName/{sciname}").json() #Get the ID number for Crassostrea virginica

commonNames = requests.get(f"https://www.marinespecies.org/rest/AphiaVernacularsByAphiaID/{aphiaID}").json() #Store the list of dictionaries outputted by this application

for name in commonNames: #Iterate through each dictionary in the list
  if name["language"] != "English":
    print(f"A common name for {sciname} in {name['language']} is {name['vernacular']}")
```

## Activity 3

For the given list of species, use the World Register of Marine Species API to determine if each species is a member of phylum Mollusca and print the result.

The API application /AphiaRecordByAphiaID/{aphiaID} returns much information about a species, including its phylum. Printing the decoded JSON output of the application would be a good first step.

```python
#Activity 3 Completion Space

sciList = ["Tristramella sacra", "Crassostrea virginica", "Mercenaria campechiensis", "Tursiops truncatus"]

```

Click "Show code" below to see a sample solution to the activity.

```python
# @title Activity 3 Solution { display-mode: "form" }
sciList = ["Tristramella sacra", "Crassostrea virginica", "Mercenaria campechiensis", "Tursiops truncatus"]

for sciname in sciList:
  aphiaID = requests.get(f"https://www.marinespecies.org/rest/AphiaIDByName/{sciname}").json()
  record = requests.get(f"https://www.marinespecies.org/rest/AphiaRecordByAphiaID/{aphiaID}").json()
  if record["phylum"] == "Mollusca":
    print(f"{sciname} is a mollusc!")
  else:
    print(f"{sciname} is not a mollusc.")
```
